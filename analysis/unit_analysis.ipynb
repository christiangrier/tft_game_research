```python
import pandas as pd
import numpy as np
import json
```

```python
file = '/Users/christiangrier/Documents/tft_game_research/tft_data/parsed_matches/100T DishsoapNA2_NA1_5433125890_20.json'
with open(file, 'r') as f:
    dishsoap = json.load(f)
# print(len(dishsoap))
data = []
for match in dishsoap:
    for units in (match['units']):
        data.append(units)
        
dishsoap_df = pd.DataFrame(data)
dishsoap_df
```

```python
champion_counts = dishsoap_df['character_id'].value_counts()
champion_counts
```

```python
# dishsoap_df['items'].value_counts()
dishsoap_df[['item_1','item_2','item_3']] = pd.DataFrame(
    dishsoap_df['items'].tolist(), 
    index=dishsoap_df.index
)
dishsoap_df.drop('items', axis=1, inplace=True)
dishsoap_df
```

```python
dishsoap_df[['item_1', 'item_2', 'item_3']].stack().value_counts()
```

```python
mask = dishsoap_df[['item_1', 'item_2', 'item_3']].notna().all(axis=1)
characters_with_3_items = dishsoap_df[mask]['character_id'].value_counts()
characters_with_3_items
```

```python
df_with_3_items = dishsoap_df[dishsoap_df[['item_1', 'item_2', 'item_3']].notna().all(axis=1)]
df_with_3_items.sort_values('character_id', ascending=True)
```
